## Loading the datasets

```python
perk = os.listdir('/content/drive/MyDrive/Parkinsons Specialization/Perkinsons') #bdf
healthy = os.listdir('/content/drive/MyDrive/Parkinsons Specialization/Healthy') #edf
normal = os.listdir('/content/drive/MyDrive/Parkinsons Specialization/Normal') #set
control = os.listdir('/content/drive/MyDrive/Parkinsons Specialization/Control') #edf
good = os.listdir('/content/drive/MyDrive/Parkinsons Specialization/Good') #bdf
```

## Defining X and Y

```python
X = np.empty((1,15000,10))
Y = []
```

## Principal Component Analysis

```python
pca = PCA(n_components=10)
```

## Preprocessing the datasets

```python
for f in healthy:
  print(f)
  raw = mne.io.read_raw_edf(os.path.join('/content/drive/MyDrive/Parkinsons Specialization/Healthy',f),preload=True)
  raw.resample(sfreq=250)
  data = raw.get_data()
  df = pd.DataFrame(data)
  df = df.transpose()
  info = df.values
  info = (info - np.mean(info))/np.std(info)
  pca = PCA(n_components=10)
  pca.fit(info)
  info = pca.transform(info)

  for i in range(0,10):
    mini = info[i*15000:15000+i*15000]
    mini = np.reshape(mini,(1,15000,10))
    X = np.concatenate((X,mini),axis=0)
    Y.append(0)

for f in control:
  raw = mne.io.read_raw_edf(os.path.join('/content/drive/MyDrive/Parkinsons Specialization/Control',f),preload=True)
  raw.resample(sfreq=250)
  data = raw.get_data()
  df = pd.DataFrame(data)
  df = df.transpose()
  info = df.values
  info = (info - np.mean(info))/np.std(info)
  pca = PCA(n_components=10)
  pca.fit(info)
  info = pca.transform(info)

  for i in range(0,1):
    mini = info[i*15000:15000+i*15000]
    mini = np.reshape(mini,(1,15000,10))
    X = np.concatenate((X,mini),axis=0)
    Y.append(0)

for f in good:
  raw = mne.io.read_raw_bdf(os.path.join('/content/drive/MyDrive/Parkinsons Specialization/Good',f),preload=True)
  raw.resample(sfreq=250)
  data = raw.get_data()
  df = pd.DataFrame(data)
  df = df.transpose()
  info = df.values
  info = (info - np.mean(info))/np.std(info)
  pca = PCA(n_components=10)
  pca.fit(info)
  info = pca.transform(info)

  for i in range(0,2):
    mini = info[i*15000:15000+i*15000]
    mini = np.reshape(mini,(1,15000,10))
    X = np.concatenate((X,mini),axis=0)
    Y.append(0)

for f in normal:
  raw = mne.io.read_raw_eeglab(os.path.join('/content/drive/MyDrive/Parkinsons Specialization/Normal',f),preload=True)
  raw.resample(sfreq=250)
  data = raw.get_data()
  df = pd.DataFrame(data)
  df = df.transpose()
  info = df.values
  info = (info - np.mean(info))/np.std(info)
  pca = PCA(n_components=10)
  pca.fit(info)
  info = pca.transform(info)

  for i in range(0,1):
    mini = info[i*15000:15000+i*15000]
    mini = np.reshape(mini,(1,15000,10))
    X = np.concatenate((X,mini),axis=0)
    Y.append(0)

for t in range(0,3):
  for f in perk:
    print(f)
    raw = mne.io.read_raw_bdf(os.path.join('/content/drive/MyDrive/Parkinsons Specialization/Perkinsons',f),preload=True)
    raw.resample(sfreq=250)
    data = raw.get_data()
    df = pd.DataFrame(data)
    df = df.transpose()
    info = df.values
    info = (info - np.mean(info))/np.std(info)
    pca = PCA(n_components=10)
    pca.fit(info)
    info = pca.transform(info)

    for i in range(0,3):
      mini = info[i*15000:15000+i*15000]
      mini = np.reshape(mini,(1,15000,10))
      X = np.concatenate((X,mini),axis=0)
      Y.append(1)



```

##  Defining the training and testing sets

```python
X = X[1:]
Y = np.array(Y)
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.1,random_state=42)
```

## Creating, compliling and running the model on the present data

```python
model = Sequential()
model.add(Convolution1D(filters=32, kernel_size=3, input_shape=[15000, 10], activation=selu))
model.add(MaxPool1D(pool_size=2))
model.add(Convolution1D(filters=64, kernel_size=3, activation=selu))
model.add(MaxPool1D(pool_size=2))
model.add(Convolution1D(filters=128, kernel_size=3, activation=selu))
model.add(MaxPool1D(pool_size=2))
model.add(Convolution1D(filters=256, kernel_size=3, activation=selu))
model.add(MaxPool1D(pool_size=2))
model.add(Convolution1D(filters=32, kernel_size=3, activation=selu))
model.add(MaxPool1D(pool_size=2))
model.add(Bidirectional(LSTM(64)))
model.add(Flatten())
model.add(Dense(640, activation=selu))
model.add(Dense(1, activation='sigmoid'))

model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])
model.summary()
# Implement learning rate scheduling
lr_scheduler = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3, verbose=1)
# Implement early stopping
early_stopping = EarlyStopping(monitor='val_loss', patience=15, restore_best_weights=True, verbose=1)

history = model.fit(X_train, Y_train, epochs=100, validation_data=(X_test, Y_test), callbacks=[lr_scheduler, early_stopping])
```

## Evaluating the model

```python
model.evaluate(X_test,Y_test)
Y_pred = np.round(model.predict(X))
print(classification_report(Y,Y_pred))
```
